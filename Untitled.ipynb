```python
age_mediana = users["age"].mean() 
users["age"] = users["age"].replace(-999, age_mediana)
users["age"].describe()
```

```python
users["city"] = users["city"].replace("?", pd.NA)
users["city"].unique()
```

```python
users['reg_date'] =  pd.to_datetime(users['reg_date'], errors="coerce")
inicio = pd.Timestamp("2022-01-01")
fin = pd.Timestamp("2024-12-31")
users.loc[(users["reg_date"] < inicio) | (users["reg_date"] > fin), "reg_date"] = np.nan
print(users["reg_date"].describe())
```

```python
usage['missing_flag'] = usage['duration'].isnull()
print(usage.groupby('missing_flag')['type'].describe())
```

```python
usage['missing_flag'] = usage['length'].isnull()
print(usage.groupby('missing_flag')['type'].describe())
```

los nulos en "duration" no entran en MAR pero en las llamadas si hay razones para creer que pertenece a un posible MAR

```python
usage["is_text"] = (usage["type"] == "text").astype(int) #conocer el total de mensajes
usage["is_call"] = (usage["type"] == "call").astype(int) #conocer el total de llamadas

usage_agg = usage.groupby("user_id").agg({"duration": "sum","is_text" : "sum","is_call" : "sum"}).reset_index()
usage_agg.head(3)
```

```python
usage_agg = usage_agg.rename(columns={
    "is_text": "cant_mensajes",
    "is_call": "cant_llamadas",
    "duration": "cant_minutos_llamadas"})
usage_agg.head(3)
```

```python
user_profile = pd.merge(usage_agg, users, on=['user_id'], how='left')
user_profile.head(5)
```

```python
# Visualizando usando BoxPlot 
columnas_numericas = ['age', 'cant_mensajes', 'cant_llamadas', 'cant_minutos_llamadas']
for col in columnas_numericas:
    plt.figure(figsize=(6,4))
    sn.boxplot(data=user_profile, y=col)
    plt.title(f"Boxplot de {col}")
    plt.show()
```

- Age: la grafica de age no muestra outliers
- cant_mensajes: aqui presentamos outliers desde los valores de 11.5 a 17.5
- cant_llamadas: la menor cantidad de outliers, con solo cuatro marcas entre 10 y 14
- cant_minutos_llamada: aqui se presenta la gran mayoria de outliers, pues tenemos una diferencia muy grande de 100 puntos

```python
columnas_limites = ['age', 'cant_mensajes', 'cant_llamadas', 'cant_minutos_llamadas']
for col in columnas_limites:
    Q1 = np.percentile(user_profile[col], 25)
    Q3 = np.percentile(user_profile[col], 75)
    IQR = Q3 - Q1
    limite_superior = Q3 + 1.5 * IQR
    print(f"limite superior = {limite_superior}")
    
```

```python
user_profile[columnas_limites].describe()
```

cant_mensajes: mantener o no outliers, porqué? si los mantendremos, aunque los valores se salen del limite superior no son suficientemente grandes para afectar el promedio 
cant_llamadas: mantener o no outliers, porqué? si los mantendremos, los valores por encima del limte no son tan grandes y permiten mantener un promedio realista
cant_minutos_llamada: mantener o no outliers, porqué? no los mantendremos, los valores por encima del limite superior son muy grandes y no sirven para obtener un promedio realista

```python

condiciones = [
    (user_profile['cant_mensajes'] < 5) & (user_profile['cant_llamadas'] < 5),
    (user_profile['cant_mensajes'] < 10) & (user_profile['cant_llamadas'] < 10)
]
valores = ['Bajo uso', 'Uso medio']
user_profile['grupo_uso'] = np.select(condiciones, valores, default='Alto uso')
```

```python
user_profile.head()
```

```python
condiciones = [
    (user_profile['age'] < 30),
    (user_profile['age'] < 60)
]
valores = ['joven', 'adulto']
user_profile['grupo_edad'] = np.select(condiciones, valores, default='adulto mayor')
```
